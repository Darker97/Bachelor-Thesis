```
import matplotlib.pyplot as plt
import matplotlib.patches as patch

import csv  
```

```
Model1_epoch = []
Model1_loss = []
Model1_val_loss = []
with open('ModelMitRauschen.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=';')
    header = next(spamreader)
    for row in spamreader:
         Wert = row[0]
         Model1_epoch.append(float(Wert))
         Wert = row[1]
         Model1_loss.append(float(Wert))
         Wert = row[2]
         Model1_val_loss.append(float(Wert))

Model1_epoch = Model1_epoch[1:]
Model1_loss = Model1_loss[1:]
Model1_val_loss = Model1_val_loss[1:]
```

```
Model2_epoch = []
Model2_loss = []
Model2_val_loss = []
with open('ModelOhneRauschen.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=';')
    header = next(spamreader)
    for row in spamreader:
         Wert = row[0]
         Model2_epoch.append(float(Wert))
         Wert = row[1]
         Model2_loss.append(float(Wert))
         Wert = row[2]
         Model2_val_loss.append(float(Wert))

Model2_epoch = Model2_epoch[1:]
Model2_loss = Model2_loss[1:]
Model2_val_loss = Model2_val_loss[1:]
```

```
plt.xkcd()
```

```
plt.plot(Model1_loss, label = 'mit Rauschen')
plt.plot(Model2_loss, label = 'ohne Rauschen')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.ylim([0, 20])
plt.legend(loc='best')
plt.show()
```

```
plt.plot(Model1_val_loss, label = 'mit Rauschen')
plt.plot(Model2_val_loss, label = 'ohne Rauschen')

plt.xlabel('Epoch')
plt.ylabel('Val_loss')
plt.ylim([0, 26])
plt.legend(loc='best')
plt.show()
```
