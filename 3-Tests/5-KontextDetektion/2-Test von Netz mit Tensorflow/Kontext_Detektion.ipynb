Zunächst verzeichnen wir die Eigenschaften des gegebenen Textes. Daraus erstellen wir einen String. Trifft die Eigenschaft zu, so setzen wir eine eins, wenn nicht eine 0. 

Dabei müsste ein Klassischer Algorithmus die Eigenschaften abgehen und daraus schlüsse ziehen. Dies ist ziemlich Aufwenig, da wir jede Mögliche Kombination abdecken müssen.

```
def Eigenschaften(testString):
  EigenschaftenArray = []
   # Hat Zahl [0-9]
  if(re.search('[0-9]', testString)):
    EigenschaftenArray.append(1)
  else:
    EigenschaftenArray.append(0)
   # Hat Buchstaben
  if(re.search('[a-zA-Z]', testString)):
    EigenschaftenArray.append(1)
  else:
    EigenschaftenArray.append(0)
   # Hat Prozentzeichen
  if(re.search('[%]', testString)):
    EigenschaftenArray.append(1)
  else:
    EigenschaftenArray.append(0)
   # Hat .
  if(re.search('[.]', testString)):
    EigenschaftenArray.append(1)
  else:
    EigenschaftenArray.append(0)
   # Hat /
  if(re.search('[/]', testString)):
    EigenschaftenArray.append(1)
  else:
    EigenschaftenArray.append(0)
   # Hat " "
  if(re.search('[ ]', testString)):
    EigenschaftenArray.append(1)
  else:
    EigenschaftenArray.append(0)
   # Länge
  EigenschaftenArray.append(len(testString))

  return np.array(EigenschaftenArray)
```

```
def createZielDaten(Number):
  Ziel = []
  for i in range(0,8):
    Ziel.append(0)
  Ziel[Number] = 1
  return Ziel
```

```
X = []
y = []

for i in TestDaten:
  X.append(Eigenschaften(i))

for i in ZielDaten:
  y.append(createZielDaten(i))

X = np.array(X)
y = np.array(y)
print(X.shape)
print(y.shape)
```

# Bauen wir das Netz

```
input = keras.layers.Input(X.shape[-1])
S1 = keras.layers.Dense(10, activation='sigmoid')(input)
Output = keras.layers.Dense(y.shape[-1], activation = 'sigmoid')(S1)


model = keras.models.Model(input, Output)

model.compile(loss='binary_crossentropy',
            optimizer='adam',
            metrics=['accuracy'])
```

```
csv_logger = CSVLogger('log.csv', append=True, separator=';')

history = model.fit(X, y, epochs=16, validation_split=0.3, callbacks=[csv_logger])
```
